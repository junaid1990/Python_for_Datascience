# Import Libraries
Seaborn automatically install these libraries

```python
# Import Libraries
import seaborn as sns
import matplotlib.pyplot as plt

# load dataset
phool = sns.load_dataset("iris")
phool

# draw a line line plot (note: line plot is always produced between 2 numeric variables)
# sns.lineplot(x="", y="", data=phool)

sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.show()
```

## Adding Limits 
- strating x or y from whihch limit

```python
# Import Libraries
import seaborn as sns
import matplotlib.pyplot as plt

# load dataset
phool = sns.load_dataset("iris")
phool

# draw a line line plot

sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.title("Phoolon ka Plot")
plt.xlim(2)
plt.ylim(1)

plt.show()
```

## Size of Figure, Change Figure Size
- i.e plt.figure(figsize= (4,2) / (8,6) / (10,8))
- stackoverflow website to setstyles

```python
# Import Libraries
import seaborn as sns
import matplotlib.pyplot as plt
# load dataset
phool = sns.load_dataset("iris")
#change figure
plt.figure(figsize=(4,2))
# draw a line line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.title("Phoolon ka Plot")
sns.set_style("dark")
plt.show()
```
