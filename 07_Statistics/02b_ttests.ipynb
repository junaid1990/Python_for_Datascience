## one sample t_Test
Test a sample with a known standard value.

**Assumptions**
- Observarion in sample is independent and indectitically distributed
- Observations in sample is mormally  distributed.

Interpretation
>- **H0**: the mean of sample is equal to knows value.
>- **H1**: the mean of sample is unequal to knows value.

```python
# one sample t-test

import seaborn as sns
import pandas as pd
from scipy.stats import ttest_1samp

# load dataset
df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
# subset df1 = df[df['age', 'sex']] what is error?
df1 = df[['age', 'sex']]
df1.head()
```

```python
# now one sample ttset matters on compare age with a known value?
df1.describe()
```

```python
# check the age and compare with a known value of 45 years old
ttest_1samp(df1['age'], 29)
```

```python
ttest_1samp(df1['age'], 14)
# data is not mormal 
```

```python
from scipy.stats import ttest_1samp
ttest_1samp(df1['age'], 50)
stat, p = ttest_1samp(df1['age'], 50)

print('stat=%.3f, p=%.3f' % (stat, p))

# adding conditional arguments for ease
if p > 0.05:
  print('Probably the same distribution')
else:
    print('Probably different distribution')
```

## Two sample t_Test
**Independent student t-test**

**Assumptions**
- Observarion in each sample are independent and indectitically distributed.
- Observations in each sample are mormally  distributed.
- observarion in each sample have the same variance

**Interpretation**
>- **H0**: the means of the samples are equal.
>- **H1**: the means of the samples are unequal.

```python
# we will compare age and fare of the male vs female passangers

df_male = df1.loc[df1['sex']=='male']
df_female = df1.loc[df1['sex']=='female']

#library
from scipy.stats import ttest_ind

stat, p = ttest_ind(df_male['age'], df_female['age'])

print('stat=%.3f, p=%.3f' % (stat, p))

# adding conditional arguments for ease
if p > 0.05:
  print('Probably the same distribution')
else:
    print('Probably different distribution')
```

## Two sample t_Test
**Paired student t-test**
Tests whether the means of two paired sampled are significantly different.

**Assumptions**
- Observarion in each sample are independent and indectitically distributed.
- Observations in each sample are mormally  distributed.
- observarion in each sample have the same variance.
- Observations across each sample are paired.

**Interpretation**
>- **H0**: the means of the samples are equal.
>- **H1**: the means of the samples are unequal.
